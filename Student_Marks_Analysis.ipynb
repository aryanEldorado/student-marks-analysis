```python
import pandas as pd

data = {
    "Name": ["Aman", "Priya", "Pratik", "Rohini", "Karan"],
    "Maths": [85,90,78,92,88],
    "Science": [80,85,75,95,89],
    "English": [78,88,72,90,85]
}

df = pd.DataFrame(data)
```

```python
df["Total"] = df[["Maths", "Science","English"]].sum(axis=1)
df["Percentage"] = df["Total"] / 3
```

```python
def grade(p):
    if p>= 90:
        return "A+"
    elif p>= 75:
        return "A"
    elif p>= 60:
        return "B"
    elif p>= 50:
        return "C"
    else:
        return "Fail"

df["Grade"] = df["Percentage"].apply(grade)
df
        
```

```python
topper = df.loc[df["Percentage"].idxmax()]
topper
```

```python
print("Topper is:", topper["Name"])
```

```python
df["Rank"] = df["Percentage"].rank(ascending=False)
df.sort_values("Rank", inplace=True)
df
```

```python
print("Math Topper:",
df.loc[df["Maths"].idxmax()]["Name"])
print("Science Topper:",
df.loc[df["Science"].idxmax()]["Name"])
print("English Topper:",
df.loc[df["English"].idxmax()]["Name"])
```

```python
df.plot(x="Name", y="Total", kind="bar",
title="Total Marks of Students")
plt.show()
```

```python
df["Grade"].value_counts().plot(kind="bar",title="Grade Distribution")
plt.show()
```

```python
import seaborn as sns

sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=1

)
plt.title("Correlation Between Subjects")
plt.show()
```
